# EDA

### Univariate Data Analysis

```python
data['price']
```

```python
# outliers
df=data[data['price']<1500]
sns.boxplot(data=df,x='price')
plt.show()
```

```python
sns.histplot(data=df,x='price',bins=100)
plt.title('Price Distribution')
plt.show()
```

```python
df.dtypes
```

```python
sns.histplot(data=df,x='availability_365')
plt.title("Availability of Rooms")
plt.show()
```

```python
sns.histplot(data=df,x='beds',bins=10)
plt.show()
```

```python
df.info()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

# Feature Engineering

```python
df.loc[:, 'price_per_bed'] = df['price'] / df['beds']
```

```python
df.head()
```

```python
df.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

# Bivariate Analysis

```python
df.dtypes
```

```python
sns.barplot(data=df,x='neighbourhood_group',y='price')
plt.show()
```

```python
sns.barplot(data=df,x='neighbourhood_group',y='price',hue='room_type')
plt.show()
```

```python
sns.pairplot(data=df,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
plt.show()
```

```python
sns.scatterplot(data=df,x='longitude',y='latitude',hue='room_type')
plt.title("Geographical Data Distribution")
plt.show()
```

```python
corr=df[['latitude','longitude','price','number_of_reviews','availability_365','beds']].corr()
corr
```
